# XML로 받아오기

- 네이버 개발자 센터에서 키 발급받고
- https://developers.naver.com/apps/#/myapps 여기서 확인
- 발급받은 키 Client ID , Client Secret 공란에다가 붙여넣기

```python
Client_ID = ''
Client_Secret = ''
```

```python
import requests as rq
import bs4
import pandas as pd
import re
import time
```

```python
# 태그의 문자열에서 html 태그를 제거하고 반환하는 함수
def getText(tag, tag_name):
    temp_str = tag.select(tag_name)[0].text.strip()
    temp_str = re.sub('<.+?>','',temp_str,0).strip()
    return temp_str

def getKeyword() :
    keyword = input('검색어를 입력해주세요 : ')
    return keyword

def connectionSite(keyword, start):
    url = '''https://openapi.naver.com/v1/search/blog.xml'''

    params = {
        'query':keyword
        ,'start':start
        ,'display':100
        ,'sort' : 'date'
    }

    headers = {
        'X-Naver-Client-Id' : Client_ID
        ,'X-Naver-Client-Secret' : Client_Secret

    }

    response = rq.get(url,params=params,headers=headers)
    soup = bs4.BeautifulSoup(response.text, 'lxml')
    # print(soup.prettify)
    return soup

def getData(soup,keyword,start_num):
    # item 태그들 가져오기
    item_list = soup.select('item')
    #print(item_list)
    for item_tag in item_list:
        # title 값 추출
        title_str = getText(item_tag, 'title')

        #링크 추출
        link_str = list(item_tag.children)[2].strip()
        #print(link_str)
        #print(list(link_str)[2])
        # description 추출
        description_str = item_tag.select('description')[0].text.strip()

        bloggername_str = getText(item_tag, 'bloggername')
        bloggerlink_str = getText(item_tag, 'bloggerlink')
        postdate_str = getText(item_tag, 'postdate')

        data_dic = {
            'title' : [title_str]
            ,'link' : [link_str]
            ,'description' : [description_str]
            ,'bloggername' : [bloggername_str]
            ,'bloggerlink' : [bloggerlink_str]
            ,'postdate' : [postdate_str]
        }

        #print(f'title : {title_str}')
        #print(f'link : {link_str}')
        #print(f'description : {description_str}')
        #print(f'bloggername : {bloggername_str}')
        #print(f'bloggerlink : {bloggerlink_str}')
        #print(f'postdate : {postdate_str}')

        df = pd.DataFrame(data_dic)
        
        global check_save
        
        if check_save == False :
            check_save = True
            df.to_csv(f'{keyword}.csv', index=False, encoding='utf-8-sig')
        else:
            df.to_csv(f'{keyword}.csv', index=False, encoding='utf-8-sig', mode='a', header=False)
        
    # total 값 확인
    total_str = soup.select('total')[0].text.strip()
        
    print(f'{start_num} / {total_str}')
        
    return int(total_str)
```

```python
# 저장했던 적이 있는지 여부...
check_save = False
start_num = 1
display_num = 100
keyword = getKeyword()
soup = connectionSite(keyword, start_num)

while True:
    #time.sleep(0.5)
    soup = connectionSite(keyword, start_num)
    total = getData(soup, keyword, start_num)
    
    start_num = start_num + 100
    
    if start_num >= 1000 or start_num >= int(total) : 
        print('마지막 입니다')
        break
    # 1000개밖에 지원 안해서..
    elif start_num == 1001:
        start_num = 1000
```

# JSON 으로 받아오기

```python
import json
```

```python
def getKeyword() :
    keyword = input('검색어를 입력해주세요 : ')
    return keyword

def connectionSite(keyword, start):
    url = '''https://openapi.naver.com/v1/search/blog.json'''

    params = {
        'query':keyword
        ,'start':start
        ,'display':100
        ,'sort' : 'date'
    }

    headers = {
        'X-Naver-Client-Id' : Client_ID
        ,'X-Naver-Client-Secret' : Client_Secret

    }

    response = rq.get(url,params=params,headers=headers)
    json_obj = json.loads(response.text)
    #soup = bs4.BeautifulSoup(response.text, 'lxml')
    # print(soup.prettify)
    return json_obj

def getData(json_obj,keyword,start_num):
    for item_tag in json_obj['items']:
        # title 값 추출
        title_str = item_tag['title']

        #링크 추출
        link_str = item_tag['link']
        #print(link_str)
        #print(list(link_str)[2])
        # description 추출
        description_str = item_tag['description']

        bloggername_str = item_tag['bloggername']
        bloggerlink_str = item_tag['bloggerlink']
        postdate_str = item_tag['postdate']

        data_dic = {
            'title' : [title_str]
            ,'link' : [link_str]
            ,'description' : [description_str]
            ,'bloggername' : [bloggername_str]
            ,'bloggerlink' : [bloggerlink_str]
            ,'postdate' : [postdate_str]
        }

        #print(f'title : {title_str}')
        #print(f'link : {link_str}')
        #print(f'description : {description_str}')
        #print(f'bloggername : {bloggername_str}')
        #print(f'bloggerlink : {bloggerlink_str}')
        #print(f'postdate : {postdate_str}')

        df = pd.DataFrame(data_dic)
        
        global check_save
        
        if check_save == False :
            check_save = True
            df.to_csv(f'{keyword}.csv', index=False, encoding='utf-8-sig')
        else:
            df.to_csv(f'{keyword}.csv', index=False, encoding='utf-8-sig', mode='a', header=False)
        
    # total 값 확인
    total_str = json_obj['total']
        
    print(f'{start_num} / {total_str}')
        
    return int(total_str)
```

```python
# 저장했던 적이 있는지 여부...
check_save = False
start_num = 1
display_num = 100
keyword = getKeyword()

while True:
    #time.sleep(0.5)
    json_obj = connectionSite(keyword, start_num)
    total = getData(json_obj, keyword, start_num)
    
    start_num = start_num + 100
    
    if start_num >= 1000 or start_num >= int(total) : 
        print('마지막 입니다')
        break
    elif start_num == 1001:
        start_num = 1000
```
